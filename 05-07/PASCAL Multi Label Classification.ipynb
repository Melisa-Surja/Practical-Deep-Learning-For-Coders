# Installing Dependencies

```python
import fastbook
fastbook.setup_book()
from fastbook import *
from fastai.vision.all import *
```

## Download Data

```python
ds_folder = Path('../../Datasets')
path = Path(ds_folder/'pascal_2007')
if not path.exists():
    path.mkdir(parents=true)    
    path = untar_data(URLs.PASCAL_2007, dest=ds_folder)  
path.ls()
```

## Prepare Dataloader

```python
# pandas Dataframe, load csv train.csv
df = pd.read_csv(path/'train.csv')
df.head()
```

```python
# Load Datablock
def get_x(r): return path/'train'/r['fname']
def get_y(r): return r['labels'].split(' ')
def splitter(df):
    train = df.index[~df['is_valid']].tolist()
    valid = df.index[df['is_valid']].tolist()
    return train,valid
db = DataBlock(blocks=(ImageBlock,MultiCategoryBlock),
               get_x=get_x, 
               get_y=get_y, 
               splitter=splitter,
               item_tfms = Resize(460),
               batch_tfms=[*aug_transforms(size=128, min_scale=0.75, mult=1.5), 
                               Normalize.from_stats(*imagenet_stats)]
              )
```

```python
# Data loaders
dls = db.dataloaders(df, num_workers=0)
dls.show_batch()
```

```python
dls.train.show_batch(max_n=10, nrows=2, unique=True)
```

## Pretrained Learner

```python
# Setup learner
learn = cnn_learner(dls,resnet18,metrics=partial(accuracy_multi,thresh=0.35))
```

```python
# Find learning rate
learn.lr_find()
```

```python
# Start training
# batch transformation and bigger image size add 13secs per epoch.
# batch transformation adds around ~6s
# bigger image size (224 instead of 128) adds around ~6s
# resnet50 is about 10-15s longer than resnet18
learn.fine_tune(7, 2e-2, freeze_epochs=4)
```

```python
# Findout best threshold
preds,targs = learn.get_preds()
```

```python
# Threshold 0.4 seems the best?
xs = torch.linspace(0.05,0.95,29)
accs = [accuracy_multi(preds, targs, thresh=i, sigmoid=False) for i in xs]
plt.plot(xs,accs);
plt.title('Threshold Curve')
plt.xlabel('Threshold')
plt.ylabel('Accuracy')
```

```python
# Test learner, top text: actual. bottom text: prediction
learn.show_results()
```
